# tutorial Seaborn

```python
import seaborn as sns
%matplotlib inline
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
data.info()
```

```python
# histograms
sns.distplot(data['total_bill'],kde=False, bins=30)
```

```python
sns.distplot(data['total_bill'],kde=True, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=data)
```

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue='sex', palette='coolwarm')
```

```python
sns.boxplot(x='day', y='total_bill',data=data, palette='rainbow')
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker',data=data, palette='coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill', data=data, hue='smoker', palette='coolwarm')
```
