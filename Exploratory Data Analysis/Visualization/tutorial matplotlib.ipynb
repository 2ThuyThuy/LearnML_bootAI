## 1. Simple Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

### 1.1. Line Colors and Style

```python
plt.plot(x, np.sin(x), color='blue') # tn mu
plt.plot(x, np.sin(x-1), color='g') # tn mu ngn (rgbcmyk)
plt.plot(x, np.sin(x-2), color='0.75') #  mu xm t 0-1
plt.plot(x, np.sin(x-3), color='#FFDD44') # mu hex code
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3)) # RGB tuple gi tr 0-1
plt.plot(x, np.sin(x-5), color='chartreuse') # mu ca HTML
```

```python
plt.plot(x, x, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

# short
plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
# linestyle and color
plt.plot(x, x, '-g') 
plt.plot(x, x+1,'--c')
plt.plot(x, x+2,'-.k')
plt.plot(x, x+3,':r')
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5) 
# ==  plt.axis([-1,11,-1.5, 1.5])
```

## 4. Density and Conrour Plots
### 4.1 visualizing a Three-Dimensional Function

```python
# z = f(x,y) = sin(x)^10 + cos(10 + y*x)*np.cos(x)
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y*x)*np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x,y) # bin i mng 1 chiu thnh mng 2 chiu
Z = f(X, Y)
Z
```

```python
plt.contour(X,Y, Z, colors='k')
```

```python
plt.contour(X,Y, Z, cmap='Set1')
```

```python
plt.contourf(X,Y, Z, 20, cmap='RdGy') # 20 khong cch u nhau
#contourf th cc ng vin c lp y
plt.colorbar()
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy')
plt.colorbar()
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
```

# 4. Histograms, Binnings, and Density

- histogram l biu  biu din phn phi ca cc bin lin tc bng cch chia nh chng thnh nhng bins c  di bng nhau v thng k s lng cc quan st ri vo tng bins

- nu bin qu nh th khng th hin r c xu hng phn phi nu ta chia cc vng ca bin qu nh. V vy chng ta c th v thm mt ng density plot th hin mt  phn phi lin lc hn bn cnh histogram

```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
plt.hist(data, bins=30, density=True, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='none')
```

```python
x1 = np.random.normal(0,0.8, 1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)
kwargs = dict(bins=30, density=True, alpha=0.5,
         histtype='stepfilled',
         edgecolor='none')
plt.hist(x1,**kwargs)
plt.hist(x2,**kwargs)
plt.hist(x3,**kwargs)
```

### Two-Dimensional Histograms and Binnings

```python
mean = [0,0]
cov =[[1,1],[1,2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
x
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
# hexagonal binnings
plt.hexbin(x, y, gridsize = 30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

# 5. Boxplots

- Biu  hp gip biu din cc i lng quan trng ca dy s nh gi tr nh nht, gi tr ln nht, t phn v(quartile), khong bin thin t phn v (interquartile Range)
- T phn v (s phn t): l i lng m t s phn b v s phn tn ca tp d liu. S phn t c 3 gi tr,  l phn t th nht (Q1), th nh (Q2), th ba (Q3). Ba gi tr ny chia mt tp hp d liu ( tp hp theo trt t t b n ln) thnh 4 phn c s lng quan st u nhau.
- Khong bin thin s phn t IQR c xc nh nh sau: IQR = Q3-Q1

- Biu  ph bin nhng khng th hin phn phi ca d liu (distribution), kh so snh cc gi tr gia cc nhm vi nhau.

# 6. Violin plots

- Violin plot l s kt hp gia biu  hp (box plot) v biu  mt  Kernel(Kernel density plot). Do vy biu  Violin va cho ta bit cc c im ca d liu nh biu  box (trung v, khong IQR,...) va cho ta bit thm v bng phn phi tng t biu  histogram

```python
data1 = np.random.normal(100, 10,  200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(70, 25, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1,data2,data3,data4]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.violinplot(data)
plt.show()
```

```python
# violin and box plot
np.random.seed(23423423)

all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))


axes[0].violinplot(all_data,showmeans=False,showmedians=True)
axes[0].set_title('Violin plot')

axes[1].boxplot(all_data)
axes[1].set_title('Box plot')

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(all_data))])
    ax.set_xlabel('Four separate samples')
    ax.set_ylabel('Observed values')
    
plt.setp(axes, xticks =[y+1 for y in range(len(all_data))], xticklabels=['x1','x2','x3','x4'])
plt.show()
```
