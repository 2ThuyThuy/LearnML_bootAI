## 1.1. Imputation using (Mean/ Median) Values:
 Tính trung bình của các giá trị không bị missing trong một cột để thay thế các giá trị missing data
 
 - Ưu điểm
     + Dễ và Nhanh
     + Tốt trên tập dữ liệu số nhỏ
 - Nhược điểm
     + Không tính đến các mối liên quan, chỉ hoạt động ở mỗi 1 cột
     + Không sử dụng được trên các đối tượng phân loại 
     + Không chính xác
     + Không tính đến sai sót trong khi gán các biến missing data
     


```python
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error
import random
import math
import numpy as np
import pandas as pd
random.seed(0)
```

```python
dataset = fetch_california_housing()
train, target = pd.DataFrame(dataset.data), pd.DataFrame(dataset.target)
train.columns = ['0','1','2','3','4','5','6','7']
train.insert(loc=len(train.columns), column = 'target', value=target)
train
```

```python
train.info()
```

```python
# Randomly replace 40% first column with NaN values
column =  train['0']
print(column.size)
missing_pct = int(column.size*0.4)
i = [random.choice(range(column.shape[0])) for _ in range(missing_pct)]
column[i] = np.NaN
print(column.shape[0] - train['0'].isna().sum())
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(strategy ='mean')
imp_train = imp_mean.fit_transform(train)
```

## 1.2 Imputation Using (Most Frequent) or (Zero/Constant) Values
làm việc với các giá trị phân loại, thay thể missing data với các giá trị xuât hiện nhiều nhất
- Ưu điểm
    + làm tốt với các giá trị phân loại
- Nhược điểm
    + Không tính đến các mối liên quan khác
    + có thể làm sai lệch dữ liệu

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(strategy='most_frequent')
imp_train = imp_mean.fit_transform(train)
```

## 1.3. Imputation Using k-NN
Thuật toán sử dụng "feature similarity" để dự đoán bất kì điểm dữ liệu mới nào

- Ưu điểm
    + Có thể dự đoán chính xác hơn phương pháp trung bình hoặc phương pháp lấy giá trị xuất hiện nhiều nhất (tùy vào dữ liệu)
- Nhược điểm
    + Tốn bộ nhớ vì KNN lưu hết dữ liệu vào trong bộ nhớ
    + K-NN nhạy cảm với các yếu tố nhiễu trong dữ liệu
