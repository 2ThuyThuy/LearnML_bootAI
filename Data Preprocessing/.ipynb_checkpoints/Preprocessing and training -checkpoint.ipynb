```python
# hàm tạo cột title từ bảng name
def make_title(s):
    try:
        ss = s[s.find(',')+2:s.find('.')]
        return ss
    except:
        return np.nan
```

```python
# chuyển các loại title về Mr, Mrs, Child
def replace_titles(sr):
    title = sr['Title']
    if sr['Age'] < 18 or title == 'Master':
        return 'Child'
    
    if title in ['Don','Major','Capt','Jonkheer','Rev','Col','Sir']:
        return 'Mr'
    elif title in ['the Countess','Mme','Dona','Lady','Mlle','Ms','Miss']:
        return 'Mrs'
    elif title == 'Dr':
        if sr['Sex'] == 'Male':
            return 'Mr'
        else :
            return 'Mrs'
    else:
        return title
```

```python
# xử lý missing cột age theo cột title
def fillna_Age(dataset):
    
    mr_data = dataset[dataset.Title == 'Mr'].copy()
    mr_data['Age'] = mr_data['Age'].fillna(mr_data[~mr_data.Age.isna()]['Age'].mean())
    
    
    mrs_data = dataset[dataset.Title == 'Mrs'].copy()
    mrs_data['Age'] = mrs_data['Age'].fillna(mrs_data[~mrs_data.Age.isna()]['Age'].mean())
    
    child_data = dataset[dataset.Title == 'Child'].copy()
    child_data['Age'] = child_data['Age'].fillna(child_data[~child_data.Age.isna()]['Age'].mean())
    
    return pd.concat([mr_data,mrs_data,child_data])
```

```python
# hàm xử lý dữ liệu chung
def Preprocessing_data(dataset):
    
    # make Title
    dataset['Title'] = dataset['Name'].apply(make_title)
    dataset['Title'] = dataset.apply(replace_titles, axis=1)
     
    #fill Age
    dataset = fillna_Age(dataset)
    
   
    dataset['Fare'].fillna(value=dataset['Fare'].mean(), inplace=True)
    dataset['Embarked'].fillna('Q', inplace=True)
    
    dataset = dataset.sort_index()
    
    dataset['Family_Size']=dataset['SibSp']+dataset['Parch']
    dataset['Fare_Per_Person']=dataset['Fare']/(dataset['Family_Size']+1)
    dataset = dataset.drop(['PassengerId','Name','Cabin','Ticket','Sex','Fare'],axis=1)
    return dataset
    
dataset_train = Preprocessing_data(dataset_train)
dataset_test = Preprocessing_data(dataset_test)
```

```python
def One_hot_columns(dataset):
    
    cat_cols = ['Embarked','Title']
    cat_cols_encoded = []
    for col in cat_cols:
        cat_cols_encoded += sorted([cat for cat in list(dataset[col].unique())])
        
    
    oh_encoder = OneHotEncoder(sparse = False,handle_unknown='ignore')
    encoder_cols = oh_encoder.fit_transform(dataset[cat_cols])
    df_new = pd.DataFrame(encoder_cols,columns = cat_cols_encoded)
    datasetnew = dataset.join(df_new)
    
    datasetnew = datasetnew.drop(cat_cols,axis=1)
    
    return datasetnew
```

```python
dataset_train = One_hot_columns(dataset_train)
dataset_test = One_hot_columns(dataset_test)
```

# test with model 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
def test_model(X_train, y_train, X_test, y_test, scaling):
    classifications = [['DecisionTreeClassifier',DecisionTreeClassifier()],
                        ['KNeighborsClassifier',KNeighborsClassifier()],
                        ['MLPClassifier',MLPClassifier(max_iter=500)]]
    Acc = pd.DataFrame(index=None,columns = ['Model', 'Accuracy','Scaling'])
    for mod in classifications:
        name = mod[0]
        model = mod[1]
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        Acc = Acc.append(pd.Series({'Model':name,
                                   'Accuracy': accuracy_score(y_test,y_pred),
                                   'Scaling': scaling}), 
                                    ignore_index=True)
    return Acc
```

```python
df1 = test_model(X_train, y_train, X_test, y_test,'No scaling')
```
